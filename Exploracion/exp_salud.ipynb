```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_td = pd.read_csv('../datos_limpios/salud/trato_directo_filtrado.csv', delimiter=';', encoding='utf-8')
df_cm = pd.read_csv('../datos_limpios/salud/convenio_marco_filtrado.csv', delimiter=';', encoding='utf-8')
df_lic = pd.read_csv('../datos_limpios/salud/licitacion_filtrado.csv', delimiter=';', encoding='utf-8')
df_ca = pd.read_csv('../datos_limpios/salud/compra_agil_filtrado.csv', delimiter=';', encoding='utf-8')
```

```python
# Agregar identificador de modalidad y combinar datasets
df_td['Modalidad'] = 'Trato Directo'
df_cm['Modalidad'] = 'Convenio Marco'
df_lic['Modalidad'] = 'Licitación'
df_ca['Modalidad'] = 'Compra Ágil'

# Combinar todos los datasets
df_combined = pd.concat([df_td, df_cm, df_lic, df_ca], ignore_index=True)

# Convertir fecha a datetime si no está ya
df_combined['FechaEnvioOC'] = pd.to_datetime(df_combined['FechaEnvioOC'], errors='coerce')

print(f"Dataset combinado: {len(df_combined):,} registros")
print(f"Modalidades: {df_combined['Modalidad'].value_counts()}")
print(f"Rango de fechas: {df_combined['FechaEnvioOC'].min()} a {df_combined['FechaEnvioOC'].max()}")
```

```python
# 1. Top 15 instituciones por gasto total
plt.figure(figsize=(14, 10))
top_instituciones = df_combined.groupby('Institucion')['MontoTotalOC'].sum().nlargest(15)

plt.barh(range(len(top_instituciones)), top_instituciones.values / 1000000, color='skyblue', edgecolor='navy', alpha=0.8)
plt.yticks(range(len(top_instituciones)), [inst[:50] + '...' if len(inst) > 50 else inst for inst in top_instituciones.index])
plt.xlabel('Monto Total (Millones CLP)', fontsize=12)
plt.title('Top 15 Instituciones de Salud por Gasto Total', fontweight='bold', fontsize=16, pad=20)
plt.grid(axis='x', alpha=0.3)

# Añadir valores en las barras
for i, v in enumerate(top_instituciones.values / 1000000):
    plt.text(v + max(top_instituciones.values) / 1000000 * 0.01, i, f'{v:,.0f}M', va='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# 2. Gasto por modalidad en top 10 instituciones
plt.figure(figsize=(16, 10))

top_10_inst = top_instituciones.head(10).index
inst_modalidad = df_combined[df_combined['Institucion'].isin(top_10_inst)]
pivot_inst = inst_modalidad.groupby(['Institucion', 'Modalidad'])['MontoTotalOC'].sum().unstack(fill_value=0)

# Crear el gráfico de barras apiladas
pivot_inst.plot(kind='bar', figsize=(16, 10), stacked=True, 
                color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4'])

plt.title('Distribución del Gasto por Modalidad\n(Top 10 Instituciones)', fontweight='bold', fontsize=16, pad=20)
plt.ylabel('Monto Total (CLP)', fontsize=12)
plt.xlabel('Institución', fontsize=12)
plt.legend(title='Modalidad', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)

# Mejorar los nombres en el eje X
ax = plt.gca()
labels = [inst[:30] + '...' if len(inst) > 30 else inst for inst in pivot_inst.index]
ax.set_xticklabels(labels)

plt.tight_layout()
plt.show()
```

```python
# 3. Gasto por región
plt.figure(figsize=(14, 8))

gasto_region = df_combined.groupby('RegionUnidadCompra')['MontoTotalOC'].sum().sort_values(ascending=False)

# Crear colores diferentes para cada barra
colors = plt.cm.Set3(range(len(gasto_region)))

bars = plt.bar(range(len(gasto_region)), gasto_region.values / 1000000, 
               color=colors, edgecolor='black', alpha=0.8)

plt.xticks(range(len(gasto_region)), gasto_region.index, rotation=45, ha='right')
plt.ylabel('Monto Total (Millones CLP)', fontsize=12)
plt.title('Gasto Público en Salud por Región', fontweight='bold', fontsize=16, pad=20)
plt.grid(axis='y', alpha=0.3)

# Añadir valores en las barras
for i, v in enumerate(gasto_region.values / 1000000):
    plt.text(i, v + max(gasto_region.values) / 1000000 * 0.01, f'{v:,.0f}M', 
             ha='center', va='bottom', fontweight='bold', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# 4. Heatmap región vs modalidad
plt.figure(figsize=(12, 10))

region_modalidad = df_combined.groupby(['RegionUnidadCompra', 'Modalidad'])['MontoTotalOC'].sum().unstack(fill_value=0)

# Crear el heatmap
sns.heatmap(region_modalidad / 1000000, 
            annot=True, 
            fmt='.0f', 
            cmap='YlOrRd',
            cbar_kws={'label': 'Monto (Millones CLP)'},
            linewidths=0.5,
            square=False)

plt.title('Distribución del Gasto por Región y Modalidad\n(Millones CLP)', 
          fontweight='bold', fontsize=16, pad=20)
plt.xlabel('Modalidad de Compra', fontsize=12)
plt.ylabel('Región', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

plt.tight_layout()
plt.show()
```

```python
# 5. Distribución porcentual por modalidad
plt.figure(figsize=(10, 8))

modalidad_counts = df_combined['Modalidad'].value_counts()
modalidad_montos = df_combined.groupby('Modalidad')['MontoTotalOC'].sum()

# Crear subplots para mostrar tanto cantidad como montos
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))

# Pie chart por número de transacciones
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4']
wedges1, texts1, autotexts1 = ax1.pie(modalidad_counts.values, labels=modalidad_counts.index, 
                                       autopct='%1.1f%%', colors=colors, startangle=90)
ax1.set_title('Distribución por Número de Transacciones', fontweight='bold', fontsize=14)

# Pie chart por monto
wedges2, texts2, autotexts2 = ax2.pie(modalidad_montos.values, labels=modalidad_montos.index, 
                                       autopct='%1.1f%%', colors=colors, startangle=90)
ax2.set_title('Distribución por Monto Total', fontweight='bold', fontsize=14)

# Mejorar el texto
for autotext in autotexts1 + autotexts2:
    autotext.set_color('white')
    autotext.set_fontweight('bold')

plt.tight_layout()
plt.show()
```

```python
# 6. Resumen estadístico
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Número de transacciones por modalidad
modalidad_counts = df_combined['Modalidad'].value_counts()
bars1 = axes[0,0].bar(modalidad_counts.index, modalidad_counts.values, 
                      color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4'])
axes[0,0].set_title('Número de Transacciones por Modalidad', fontweight='bold')
axes[0,0].set_ylabel('Número de Transacciones')
axes[0,0].tick_params(axis='x', rotation=45)

# Añadir valores en las barras
for bar in bars1:
    height = bar.get_height()
    axes[0,0].text(bar.get_x() + bar.get_width()/2., height + height*0.01,
                   f'{int(height):,}', ha='center', va='bottom', fontweight='bold')

# Monto promedio por modalidad
modalidad_promedio = df_combined.groupby('Modalidad')['MontoTotalOC'].mean()
bars2 = axes[0,1].bar(modalidad_promedio.index, modalidad_promedio.values / 1000, 
                      color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4'])
axes[0,1].set_title('Monto Promedio por Modalidad', fontweight='bold')
axes[0,1].set_ylabel('Monto Promedio (Miles CLP)')
axes[0,1].tick_params(axis='x', rotation=45)

# Añadir valores en las barras
for bar in bars2:
    height = bar.get_height()
    axes[0,1].text(bar.get_x() + bar.get_width()/2., height + height*0.01,
                   f'{int(height):,}K', ha='center', va='bottom', fontweight='bold')

# Top 5 regiones por transacciones
region_counts = df_combined['RegionUnidadCompra'].value_counts().head(5)
axes[1,0].bar(range(len(region_counts)), region_counts.values, color='lightcoral')
axes[1,0].set_title('Top 5 Regiones por Número de Transacciones', fontweight='bold')
axes[1,0].set_ylabel('Número de Transacciones')
axes[1,0].set_xticks(range(len(region_counts)))
axes[1,0].set_xticklabels(region_counts.index, rotation=45, ha='right')

# Estadísticas generales
total_gastos = df_combined['MontoTotalOC'].sum()
num_transacciones = len(df_combined)
num_instituciones = df_combined['Institucion'].nunique()
num_proveedores = df_combined['Proveedor'].nunique()

stats_text = f"""
ESTADÍSTICAS GENERALES

💰 Gasto Total: ${total_gastos:,.0f} CLP
📊 Transacciones: {num_transacciones:,}
🏢 Instituciones: {num_instituciones:,}
🏪 Proveedores: {num_proveedores:,}

💵 Promedio por transacción: 
   ${total_gastos/num_transacciones:,.0f} CLP

📈 Monto por institución:
   ${total_gastos/num_instituciones:,.0f} CLP

💼 Monto por proveedor:
   ${total_gastos/num_proveedores:,.0f} CLP
"""

axes[1,1].text(0.1, 0.5, stats_text, transform=axes[1,1].transAxes, fontsize=12, 
               verticalalignment='center', fontfamily='monospace',
               bbox=dict(boxstyle="round,pad=0.5", facecolor="lightblue", alpha=0.8))
axes[1,1].set_title('Resumen Estadístico', fontweight='bold')
axes[1,1].axis('off')

plt.tight_layout()
plt.show()
```

```python
# ====================================================================
# 2. CONCENTRACIÓN EN PROVEEDORES
# ====================================================================

# Top 20 proveedores por ventas
plt.figure(figsize=(16, 12))
top_proveedores = df_combined.groupby('Proveedor')['MontoTotalOC'].sum().nlargest(20)

plt.barh(range(len(top_proveedores)), top_proveedores.values / 1000000, 
         color='darkorange', edgecolor='darkred', alpha=0.8, height=0.7)

plt.yticks(range(len(top_proveedores)), 
           [prov[:50] + '...' if len(prov) > 50 else prov for prov in top_proveedores.index],
           fontsize=12)

plt.xlabel('Monto Total (Millones CLP)', fontsize=16, fontweight='bold')
plt.title('TOP 20 PROVEEDORES POR VENTAS AL SECTOR SALUD', 
          fontsize=18, fontweight='bold', pad=30)

# Añadir valores en las barras
for i, v in enumerate(top_proveedores.values / 1000000):
    plt.text(v + max(top_proveedores.values) / 1000000 * 0.02, i, 
             f'{v:,.0f}M', va='center', fontweight='bold', fontsize=11)

plt.grid(axis='x', alpha=0.4, linestyle='--')
plt.tight_layout()
plt.subplots_adjust(left=0.35)
plt.show()
```

```python
# Análisis de concentración (Pareto) - Proveedores
plt.figure(figsize=(14, 10))

total_gastos = df_combined['MontoTotalOC'].sum()
proveedores_ordenados = df_combined.groupby('Proveedor')['MontoTotalOC'].sum().sort_values(ascending=False)
proveedores_ordenados_pct = (proveedores_ordenados.cumsum() / total_gastos * 100)

# Encontrar puntos clave
punto_50 = (proveedores_ordenados_pct <= 50).sum()
punto_80 = (proveedores_ordenados_pct <= 80).sum()
punto_95 = (proveedores_ordenados_pct <= 95).sum()

# Crear el gráfico
plt.plot(range(len(proveedores_ordenados_pct)), proveedores_ordenados_pct.values, 
         linewidth=3, color='darkblue')

# Líneas de referencia
plt.axhline(y=50, color='green', linestyle='--', alpha=0.7, label=f'{punto_50} proveedores = 50% del gasto')
plt.axhline(y=80, color='orange', linestyle='--', alpha=0.7, label=f'{punto_80} proveedores = 80% del gasto')
plt.axhline(y=95, color='red', linestyle='--', alpha=0.7, label=f'{punto_95} proveedores = 95% del gasto')

plt.axvline(x=punto_50, color='green', linestyle='--', alpha=0.7)
plt.axvline(x=punto_80, color='orange', linestyle='--', alpha=0.7)
plt.axvline(x=punto_95, color='red', linestyle='--', alpha=0.7)

plt.xlabel('Número de Proveedores', fontsize=14, fontweight='bold')
plt.ylabel('% Acumulado del Gasto', fontsize=14, fontweight='bold')
plt.title('ANÁLISIS DE CONCENTRACIÓN DE PROVEEDORES (Pareto)', 
          fontsize=16, fontweight='bold', pad=20)

plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Mostrar estadísticas
print(f"📊 CONCENTRACIÓN DE PROVEEDORES:")
print(f"  • {punto_50:,} proveedores controlan el 50% del gasto")
print(f"  • {punto_80:,} proveedores controlan el 80% del gasto")
print(f"  • {punto_95:,} proveedores controlan el 95% del gasto")
print(f"  • Total de proveedores: {len(proveedores_ordenados):,}")
```

```python
# Número de contratos por proveedor vs monto
plt.figure(figsize=(16, 10))

prov_stats = df_combined.groupby('Proveedor').agg({
    'MontoTotalOC': ['sum', 'count']
}).reset_index()
prov_stats.columns = ['Proveedor', 'MontoTotal', 'NumContratos']

# Filtrar solo proveedores con múltiples contratos
prov_multi = prov_stats[prov_stats['NumContratos'] >= 5].copy()

# Crear scatter plot
scatter = plt.scatter(prov_multi['NumContratos'], prov_multi['MontoTotal'] / 1000000, 
                     alpha=0.6, s=60, c=prov_multi['MontoTotal'] / 1000000, 
                     cmap='viridis', edgecolors='black')

plt.xlabel('Número de Contratos', fontsize=14, fontweight='bold')
plt.ylabel('Monto Total (Millones CLP)', fontsize=14, fontweight='bold')
plt.title('RELACIÓN: NÚMERO DE CONTRATOS vs MONTO TOTAL\n(Proveedores con 5+ contratos)', 
          fontsize=16, fontweight='bold', pad=20)

# Añadir colorbar
cbar = plt.colorbar(scatter)
cbar.set_label('Monto Total (Millones CLP)', fontsize=12)

plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# ====================================================================
# 4. DETECCIÓN DE SOBREPRECIOS
# ====================================================================

# Análisis por rubros
plt.figure(figsize=(16, 12))

if 'RubroN1' in df_combined.columns:
    # Top 10 rubros por gasto
    top_rubros = df_combined.groupby('RubroN1')['MontoTotalOC'].sum().nlargest(10).index
    df_rubros = df_combined[df_combined['RubroN1'].isin(top_rubros) & 
                           (df_combined['MontoTotalOC'] > 0)]
    
    # Crear boxplot
    plt.figure(figsize=(16, 10))
    box_plot = plt.boxplot([df_rubros[df_rubros['RubroN1'] == rubro]['MontoTotalOC'].values 
                           for rubro in top_rubros],
                          labels=[rubro[:25] + '...' if len(rubro) > 25 else rubro 
                                 for rubro in top_rubros],
                          patch_artist=True)
    
    # Colorear las cajas
    colors = plt.cm.Set3(range(len(top_rubros)))
    for patch, color in zip(box_plot['boxes'], colors):
        patch.set_facecolor(color)
        patch.set_alpha(0.7)
    
    plt.yscale('log')
    plt.ylabel('Monto Total (CLP - Escala Log)', fontsize=14, fontweight='bold')
    plt.xlabel('Rubro', fontsize=14, fontweight='bold')
    plt.title('DISTRIBUCIÓN DE MONTOS POR RUBRO\n(Top 10 Rubros - Escala Logarítmica)', 
              fontsize=16, fontweight='bold', pad=20)
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.show()
```

```python
# Histograma de montos con detección de outliers
plt.figure(figsize=(16, 10))

# Calcular estadísticas
Q1 = df_combined['MontoTotalOC'].quantile(0.25)
Q3 = df_combined['MontoTotalOC'].quantile(0.75)
IQR = Q3 - Q1
limite_superior = Q3 + 1.5 * IQR
outliers = df_combined[df_combined['MontoTotalOC'] > limite_superior]

# Crear histograma
plt.hist(df_combined['MontoTotalOC'], bins=100, alpha=0.7, color='skyblue', 
         edgecolor='black', density=True)

# Línea de outliers
plt.axvline(x=limite_superior, color='red', linestyle='--', linewidth=2, 
            label=f'Límite outliers: ${limite_superior:,.0f}')

plt.xlabel('Monto Total (CLP)', fontsize=14, fontweight='bold')
plt.ylabel('Densidad', fontsize=14, fontweight='bold')
plt.title('DISTRIBUCIÓN DE MONTOS TOTALES\n(Detección de valores atípicos)', 
          fontsize=16, fontweight='bold', pad=20)
plt.xscale('log')
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"📊 DETECCIÓN DE OUTLIERS:")
print(f"  • Número de outliers: {len(outliers):,}")
print(f"  • Porcentaje de outliers: {len(outliers)/len(df_combined)*100:.2f}%")
print(f"  • Monto promedio outliers: ${outliers['MontoTotalOC'].mean():,.0f}")
```

```python
# Actividad económica de proveedores
plt.figure(figsize=(16, 12))

if 'ActividadProveedor' in df_combined.columns:
    top_actividades = df_combined.groupby('ActividadProveedor')['MontoTotalOC'].sum().nlargest(15)
    
    plt.barh(range(len(top_actividades)), top_actividades.values / 1000000, 
             color='lightcoral', alpha=0.8, height=0.7)
    
    # Función para dividir texto largo
    def format_label(text, max_length=35):
        if len(text) <= max_length:
            return text
        # Buscar un espacio cerca del medio
        mid = max_length // 2
        space_pos = text.find(' ', mid)
        if space_pos == -1 or space_pos > max_length:
            return text[:max_length-3] + '...'
        return text[:space_pos] + '\n' + text[space_pos+1:]
    
    labels = [format_label(act) for act in top_actividades.index]
    
    plt.yticks(range(len(top_actividades)), labels, fontsize=10)
    
    plt.xlabel('Monto Total (Millones CLP)', fontsize=14, fontweight='bold')
    plt.title('TOP 15 ACTIVIDADES ECONÓMICAS DE PROVEEDORES', 
              fontsize=16, fontweight='bold', pad=20)
    
    # Añadir valores
    for i, v in enumerate(top_actividades.values / 1000000):
        plt.text(v + max(top_actividades.values) / 1000000 * 0.02, i, 
                 f'{v:,.0f}M', va='center', fontweight='bold', fontsize=11)
    
    plt.grid(axis='x', alpha=0.3)
    plt.tight_layout()
    plt.subplots_adjust(left=0.45)
    plt.show()
```
